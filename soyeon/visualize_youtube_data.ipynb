```python
youtubers = channel_data['이름'].values.tolist()
subscribers = list(map(int, channel_data['구독자수'].values.tolist()))
video_count = list(map(int, channel_data['동영상개수'].values.tolist()))
view_count = list(map(int, channel_data['동영상총조회수'].values.tolist()))
```

```python
import re

def preprocess_df(df):

    # T, Z 지우기
    df['uploaded_date'] = [re.sub('[A-Z]'," ",s) for s in df['uploaded_date']]
    # Z 공백 제거
    df['uploaded_date'] = df['uploaded_date'].str.rstrip()

    # 포맷 바꾸기
    df['uploaded_date'] =pd.to_datetime(df['uploaded_date'],
                                      format='%Y-%m-%d', errors='raise')
    df['like_count'] = pd.to_numeric(df['like_count'], errors='coerce').fillna(0).astype(int)

    df['date'] = df['uploaded_date'].dt.date

    # datetime - index설정  (resample사용하기 위해 - 안그러면 에러나옴)
    df = df.set_index('uploaded_date')

    return df
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.rcParams['font.family'] = 'AppleGothic'

channel_df = pd.read_csv('./../src/TB_CHANNEL_INFO.csv', encoding='utf-8', index_col=0)

category_dict = {'ANIMAL': '동물', 'FILM': '영화', 'SPORT': '스포츠',
                 'GAMING': '게임', 'MUKBANG': '먹방'}


def get_youtuber_rank_graph(category, category_name):
    df = channel_df[channel_df['category'] == category]

    if category == 'FILM':
        category = 'MOVIE'
    elif category == 'SPORT':
        category = 'SPORTS'
    elif category == 'GAMING':
        category = 'GAME'

    # 총 조회수 / 총 비디오 개수
    df['total_views_per_videos'] = (df['total_views'] // df['videos_count']) // 100000

    # 구독자 수 / 십만
    df['subscribers_count_100000'] = df['subscribers_count'] // 100000

    w = 0.15
    nrow = df.shape[0] # 행의 갯수
    idx = np.arange(nrow)

    fig, ax1 = plt.subplots(figsize = (20, 10))

    # 첫번째 y축: Subscribers와 Videos
    ax1.bar(idx - 2 * w, df['subscribers_count_100000'], width = w, label='구독자 수(단위: 십만)', color='#aec7e8')
    ax1.bar(idx - w, df['total_views_per_videos'], width = w, label='평균 조회 수(단위: 십만)', color='#FF69B4')
    ax1.set_ylabel('구독자 수(단위: 백만) / 평균 조회 수(단위: 십만)', fontsize=12)

    # 두번째 y축: Total Views
    ax2 = ax1.twinx()
    ax2.bar(idx, df['videos_count'] / 100, width = w, label='업로드 빈도율(단위: %)', color='#00FF7F')
    ax2.set_ylabel('업로드 빈도율(단위 : %)', fontsize=12)

    # x축, 타이틀, 범례
    plt.xticks(idx, df['channel_name'], rotation = 30)
    plt.title(category_name +' 유튜버 순위', fontsize=16)
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')

    # y축 범위


    ax1_max = max(df['subscribers_count_100000'].max(),df['total_views_per_videos'].max())

    ax1_min = min(df['subscribers_count_100000'].min(),df['total_views_per_videos'].min())
    ax1_range = ax1_max - ax1_min
    ax1.set_ylim([0, ax1_max + 0.1 * ax1_range])

    ax2.set_ylim([0, df['videos_count'].max() / 100])

    plt.savefig('./../src/graph/{0}/{0}_CHANNEL_YOUTUBER_GRAPH.png'.format(category),
                facecolor='#ffffff',
                edgecolor='black',
                format='png', dpi=200)
    plt.show()
```

```python
import re

def preprocess_df(df):

    # T, Z 지우기
    df['uploaded_date'] = [re.sub('[A-Z]'," ",s) for s in df['uploaded_date']]
    # Z 공백 제거
    df['uploaded_date'] = df['uploaded_date'].str.rstrip()

    # 포맷 바꾸기
    df['uploaded_date'] =pd.to_datetime(df['uploaded_date'],
                                      format='%Y-%m-%d', errors='raise')
    df['like_count'] = pd.to_numeric(df['like_count'], errors='coerce').fillna(0).astype(int)

    df['date'] = df['uploaded_date'].dt.date
    # datetime - index설정  (resample사용하기 위해 - 안그러면 에러나옴)
    df = df.set_index('uploaded_date')

    return df
```
