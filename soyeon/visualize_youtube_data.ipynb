# 먹방 유튜버 8인의 구독자, 동영상, 조회수, 구독자 대비 조회율 그래프 그리기

```python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
```

```python
plt.rcParams['font.family'] = 'AppleGothic'
```

```python
channel_data = pd.read_csv('먹방_유튜버_8인_채널_정보.csv')

channel_data
```

```python
youtubers = channel_data['이름'].values.tolist()
subscribers = list(map(int, channel_data['구독자수'].values.tolist()))
video_count = list(map(int, channel_data['동영상개수'].values.tolist()))
view_count = list(map(int, channel_data['동영상총조회수'].values.tolist()))
```

```python
views_per_subscribers = [int(s) / int(v) for s, v in zip(subscribers, view_count)]

views_per_subscribers
```

```python
fig, ax = plt.subplots(2, 2, figsize=(30, 20))

# 유튜버별 색상 리스트 설정
colors = ['orangered', 'orange', 'yellow', 'greenyellow', 'palegreen',
          'turquoise', 'lightskyblue', 'blueviolet']

# 구독자 수 그래프
ax[0, 0].bar(youtubers, subscribers, color=colors)
ax[0, 0].set_title('Subscribers')

# 동영상 개수 그래프
ax[0, 1].bar(youtubers, video_count, color=colors)
ax[0, 1].set_title('Video Count')

# 동영상 총 조회수 그래프
ax[1, 0].bar(youtubers, view_count, color=colors)
ax[1, 0].set_title('View Count')


# 각 유튜브 별 구독자 수 대비 = 동영상 총 조회수 / 구독자 수
ax[1, 1].bar(youtubers, views_per_subscribers, color=colors)
ax[1, 1].set_title('Views per Subscribers')
```

# 먹방 유튜브 8인의 최근 동영상 데이터 100개의 그래프 그려보기

```python
video_data = pd.read_csv('먹방_유튜버_8인_최근동영상100개_상세정보.csv', encoding='utf-8', index_col=0)
```

```python
video_data
```

```python
video_data.info()
```

```python
# 최근 동영상 데이터 100개의 총 조회수
total_recent_view_count = video_data.groupby("이름")['조회수'].sum()
total_recent_view_count_df = total_recent_view_count.to_frame()
```

```python
name_list = total_recent_view_count_df.index.tolist()
name_list
```

```python
value_list = [value[0] for value in total_recent_view_count_df.values.tolist()]
value_list
```

```python
# 유튜버별 색상 리스트 설정
colors = ['orangered', 'orange', 'yellow', 'greenyellow', 'palegreen',
          'turquoise', 'lightskyblue', 'blueviolet']

# 막대 그래프 시각화
plt.figure(figsize=(30, 20))
plt.rc('font', size=20)

plt.barh(name_list, value_list, color=colors)
plt.xlabel('유튜버')
plt.ylabel('조회수')
plt.title('최근 동영상 100개의 총 조회수')
plt.show()
```
