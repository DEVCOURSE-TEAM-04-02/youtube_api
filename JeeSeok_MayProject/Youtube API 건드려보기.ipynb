```python
def extract_playlist_videos_info(channel):   #재생목록으로 영상 조회하기
    search_response = youtube.search().list(
        q = channel,
        order = "relevance", # 관련성 순
        part = "snippet",
        maxResults = 50   # 최대 50개의 결과 접근
    ).execute()

    channel_ID = search_response['items'][0]['id']['channelId']

    playlists=youtube.playlists().list(
        channelId=channel_ID,
        part='snippet',
        maxResults=25).execute()      # playlist 접근
        
    for i in playlists['items']:
        if '리그오브레전드' in i['snippet']['title'] or '롤' in i['snippet']['title']:
            playlist_id = i['id']
            break                     # 리그오브레전드 재생목록 찾기

    playlist_videos=youtube.playlistItems().list(
        playlistId=playlist_id,
        part='snippet',
        maxResults=100).execute()      # 리그오브레전드 재생목록 접근
    
    video_Ids = []
    for j in playlist_videos['items']:
        if j['snippet']['videoOwnerChannelTitle'] != channel:
            continue
        video_Ids.append(j['snippet']['resourceId']['videoId'])  #재생목록의 영상 ID 추출
    video_dict = {}
    for k in range(0,len(video_Ids)):
        videoIdslists = youtube.videos().list(
            part='snippet, statistics',
            id=video_Ids[k],
        ).execute()
        
        items = videoIdslists['items'][0]
        try :
            likeCount = items['statistics']['likeCount']    # 좋아요 수 예외처리
        except KeyError as e:
            likeCount = "표시되지 않음"
            
        try :
            commentCount = items['statistics']['commentCount']    # 댓글수 예외처리
        except KeyError as e:
            commentCount = "표시되지 않음"
            
        video_dict[video_Ids[k]] = [items['snippet']['title'],items['snippet']['publishedAt'],items['statistics']['viewCount'],
                                    likeCount,commentCount]

        
    video_df = pd.DataFrame.from_dict(data=video_dict, orient='index',columns=['제목','업로드날짜','조회수','좋아요','댓글'])
    
    return video_df
```

# 관련성 기준

```python
search_response = youtube.search().list(
    q = f"#리그오브레전드",
    order = "relevance", # 관련성 순
    regionCode = "KR",   # 한국 게시물
    type = 'video',      # 비디오만 
    videoDuration = 'medium', # 4~20분 영상
    part = "snippet",
    maxResults = 50   # 최대 50개의 결과 접근
    ).execute()
```

```python
pprint.pprint(search_response)
```

```python
rel_videoIds = []
for i in search_response['items']:
    rel_videoIds.append(i["id"]['videoId'])    
```

```python
rel_video_dicts = {}
for k in range(0,len(search_response['items'])):
    videoIdslists = youtube.videos().list(
        part='snippet, statistics',
        id=rel_videoIds[k],
    ).execute()
    items = videoIdslists['items'][0]
    rel_video_dicts[rel_videoIds[k]] = {"채널명" : items['snippet']['channelTitle'],
                               "영상 제목" : items['snippet']['title'],
                               "조회수" : items['statistics']['viewCount'],
                               "좋아요수" : items['statistics']['likeCount'],
                               "댓글수" : items['statistics']['commentCount']}
```
