```python
def extract_channelId(channel_name):
    #타입을 채널로 설정하여 채널 이름으로 검색했을 때 첫번째 나오는 채널의 ID를 갖고온다.
    
    search_response = youtube.search().list(
        q = channel_name,
        order = "relevance", # 관련성 순
        type = 'channel',
        part = "snippet",
        maxResults = 2,
        ).execute()

    return search_response['items'][0]['id']['channelId']
```

```python
def extract_channel_info(channel_id):
    # 채널 id로 접근하여 채널의 총 조회수, 구독자수, 영상 개수를 갖고온다.
    
    channel_response = youtube.channels().list(
        part="statistics",
        id=channel_id
    ).execute()

    total_viewCount = channel_response['items'][0]['statistics']['viewCount']
    subscriberCount = channel_response['items'][0]['statistics']['subscriberCount']
    videoCount = channel_response['items'][0]['statistics']['videoCount']
    return [total_viewCount, subscriberCount, videoCount]
```

```python
def extract_video_Id(channel_Id):
    # 채널 Id로 접근하여 채널의 최근 100개 동영상의 ID 추출하기

    video_Ids = []

    videos_response = youtube.search().list(
        part="snippet",
        channelId=channel_Id,
        order="date",
        type="video",
        maxResults=50
    ).execute()
    
    nextPageToken = videos_response['nextPageToken']
    
    for video in videos_response['items']:
        video_Ids.append(video['id']['videoId'])
    
    # 한번에 50개만 나오기 때문에 다음 페이지로 넘어가야 함.
    nextpage_response = youtube.search().list(
        part="snippet",
        channelId=channel_Id,
        order="date",
        type="video",
        pageToken = nextPageToken,
        maxResults=50).execute()
    
    for video in nextpage_response['items']:
        if video['id']['videoId'] in video_Ids:
            continue
        video_Ids.append(video['id']['videoId'])

    return video_Ids
```

```python
def extract_video_info(videoId):
    view_video_dicts = {}
    for k in range(0,len(videoId)):
        videoIdslists = youtube.videos().list(
            part='snippet, statistics',
            id=videoId[k],
        ).execute()
        items = videoIdslists['items'][0]
        try :
            likeCount = items['statistics']['likeCount']
        except KeyError as e:
            likeCount = "표시되지 않음"
            
        try :
            commentCount = items['statistics']['commentCount']
        except KeyError as e:
            commentCount = "표시되지 않음"
            
        try :
            tags = items['snippet']['tags']
        except KeyError as e:
            tags = ""

        view_video_dicts[videoId[k]] = [items['snippet']['channelId'], 
                                        items['snippet']['title'], items['snippet']['publishedAt'], 
                                        items['statistics']['viewCount'], likeCount,
                                        commentCount, tags]
        video_df = pd.DataFrame.from_dict(data=view_video_dicts, orient='index',
                                          columns=['채널Id','영상제목','업로드날짜','조회수','좋아요','댓글','해시태그'])
    return video_df
```

```python
def extract_playlist_videos_info(channel):   #재생목록으로 영상 조회하기
    search_response = youtube.search().list(
        q = channel,
        order = "relevance", # 관련성 순
        part = "snippet",
        maxResults = 10   # 최대 10개의 결과 접근
    ).execute()

    channel_ID = search_response['items'][0]['id']['channelId']

    playlists=youtube.playlists().list(
        channelId=channel_ID,
        part='snippet',
        maxResults=25).execute()      # playlist 접근
        
    for i in playlists['items']:
        if '리그오브레전드' in i['snippet']['title'] or '롤' in i['snippet']['title']:
            playlist_id = i['id']
            break                     # 리그오브레전드 재생목록 찾기

    playlist_videos=youtube.playlistItems().list(
        playlistId=playlist_id,
        part='snippet',
        maxResults=200).execute()      # 리그오브레전드 재생목록 접근
    
    video_Ids = []
    for j in playlist_videos['items']:
        try :
            owner = j['snippet']['videoOwnerChannelTitle']
        except KeyError as e:
            pass
        if owner != channel:
            continue
        video_Ids.append(j['snippet']['resourceId']['videoId'])  #재생목록의 영상 ID 추출
        
    video_dict = {}
    for k in range(0,len(video_Ids)):
        videoIdslists = youtube.videos().list(
            part='snippet, statistics',
            id=video_Ids[k],
        ).execute()
        
        items = videoIdslists['items'][0]
        try :
            likeCount = items['statistics']['likeCount']    # 좋아요 수 예외처리
        except KeyError as e:
            likeCount = "표시되지 않음"
            
        try :
            commentCount = items['statistics']['commentCount']    # 댓글수 예외처리
        except KeyError as e:
            commentCount = "표시되지 않음"
            
        video_dict[video_Ids[k]] = [items['snippet']['title'],items['snippet']['publishedAt'],items['statistics']['viewCount'],
                                    likeCount,commentCount]

        
    video_df = pd.DataFrame.from_dict(data=video_dict, orient='index',columns=['제목','업로드날짜','조회수','좋아요','댓글'])
    
    return video_df
```
