```python
# 채널 정보 넣기

channel = pd.read_csv('https://raw.githubusercontent.com/DEVCOURSE-TEAM-04-02/youtube_api/main/src/%EB%A8%B9%EB%B0%A9_%EC%9C%A0%ED%8A%9C%EB%B2%84_8%EC%9D%B8_%EC%B1%84%EB%84%90_%EC%A0%95%EB%B3%B4.csv')

for i in range(len(channel)):
    data = channel.loc[i]
    
    if "\'" in data['channel_name']:  # 채널이름에 '가 있으면 안들어가지더라구여.. ㅠㅠ
        channel_name = data['channel_name'].replace("\'", "")
    
    sql = "insert into TB_CHANNEL_INFO values('" + data['category'] + "','" +  data['channel_id'] + "','" +  \
          channel_name + "','" +  data['channel_logo'] + "','" +  str(data['subscribers_count']) + "','" + \
          str(data['videos_count']) + "','"  + str(data['total_views']) + "');"
    
    try:
        cur.execute(sql)
        con.commit()
    except:
        con.rollback()  
        
con.close()
```
